```python
import pandas as pd
```

```python
df = pd.read_excel('tabela_cidades_uf_ibge.xlsx')
df.head()
```

```python
print('\n RANKING DAS CIDADES COM NOMES REPETIDOS MAIS DE 2 VEZES\n')
nomes_iguais = df.NM_CIDADE.value_counts().reset_index()
nomes_iguais.columns = ['NM_CIDADE','QTDE']
nomes_iguais.query("QTDE > 2")
```

```python
print('\n LOCALIZAÇÃO DAS DUAS CIDADES COM O NOME MAIS REPETIDO \n')
display(df.query("NM_CIDADE=='Bom Jesus'")[['NM_CIDADE', 'NM_UF']])
display(df.query("NM_CIDADE=='São Domingos'")[['NM_CIDADE', 'NM_UF']])
```

```python
# Consultando as localizações das Duas printipais cidades
Bom_Jesus = df[df['NM_CIDADE'].str.contains('Bom Jesus', case=False)]['NM_CIDADE'].tolist()
SaoDomingos = df[df['NM_CIDADE'].str.contains('São Domingos', case=False)]['NM_CIDADE'].tolist()
print('\n LOCALIZAÇÃO DAS CIDADES QUE CONTEM EM SEUS NOMES, OS NOMES MAIS REPETIDOS. \n')
display(df.query(f"NM_CIDADE=={Bom_Jesus}")[['NM_CIDADE', 'NM_UF']])
display(df.query(f"NM_CIDADE=={SaoDomingos}")[['NM_CIDADE', 'NM_UF']])
```

```python
## Outras Curiosidades: 
Jesus = df[df['NM_CIDADE'].str.contains('Jesus', case=False)]['NM_CIDADE'].tolist()
print("\n LISTA DE CIDADES QUE TEM CONTÊM O NOME 'JESUS'.\n")
display(df.query(f"NM_CIDADE=={Jesus}")[['NM_CIDADE', 'NM_UF']])
```
